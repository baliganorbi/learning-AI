```python
from dotenv import load_dotenv

import os

load_dotenv()

PIPELINE_ID = "ce09a8e4-a411-4d82-abde-b8b2ca6c6fb4"
LLAMACLOUD_API_KEY = os.getenv("LLAMACLOUD_API_KEY")
```

```python
from llama_cloud import AsyncLlamaCloud

client = AsyncLlamaCloud(api_key=LLAMACLOUD_API_KEY)
question = input("Enter your question: ")

# Retrieve relevant nodes from the index
results = await client.pipelines.retrieve(
  pipeline_id=PIPELINE_ID,
  query=question,
  # -- Customize search behavior --
  # dense_similarity_top_k=20,
  # sparse_similarity_top_k=20,
  # alpha=0.5,
  # -- Control reranking behavior --
  # enable_reranking=True,
  # rerank_top_n=5,
)

for n in results.retrieval_nodes:
  print(f"Score: {n.score}, Text: {n.node.text}")
```
